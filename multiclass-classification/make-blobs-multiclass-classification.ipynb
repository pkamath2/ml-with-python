```python
import sklearn
from sklearn.datasets import make_blobs
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt 
import numpy as np
```

```python
X,y = make_blobs(random_state=42)
```

```python
linearSVM = sklearn.svm.LinearSVC().fit(X,y)
print("Coefficient:\n{}".format(linearSVM.coef_))
print("Intercept:\n{}".format(linearSVM.intercept_))
```

```python
line = np.linspace(-15,15)
plt.scatter(X[:,0],X[:,1],c=y)
plt.xlim(-15,15)
plt.ylim(-15,15)
for coef, intercept, color in zip(linearSVM.coef_, linearSVM.intercept_,['b','r','g']):
    plt.plot(line,-(line*coef[0]+intercept)/coef[1],c=color)
```
