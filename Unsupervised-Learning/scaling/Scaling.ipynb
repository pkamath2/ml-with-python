```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVC
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer['data'],cancer['target'], random_state=42)
print(cancer['data'])
print(cancer['target'])
```

```python
#Scale the training set
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
print("Transformed shape:\n{}".format(X_train_scaled.shape))
print("per-feature min before scaling:{}".format(X_train.min(axis=0)))
print("per-feature max before scaling:{}".format(X_train.max(axis=0)))
print("per-feature min after scaling:{}".format(X_train_scaled.min(axis=0)))
print("per-feature max after scaling:{}".format(X_train_scaled.max(axis=0)))
```

```python
#Scale the test set
scaler = MinMaxScaler()
scaler.fit(X_test)
X_test_scaled = scaler.transform(X_test)
print("Transformed shape:\n{}".format(X_test_scaled.shape))
print("per-feature min before scaling:{}".format(X_test.min(axis=0)))
print("per-feature max before scaling:{}".format(X_test.max(axis=0)))
print("per-feature min after scaling:{}".format(X_test_scaled.min(axis=0)))
print("per-feature max after scaling:{}".format(X_test_scaled.max(axis=0)))
```

```python
fig, axes = plt.subplots(1,2,figsize=(13,4))
axes[0].scatter(X_train[:,0],X_train[:,1],c=y_train,label="Training set",s=60)
axes[0].scatter(X_test[:,0],X_test[:,1],c=y_test,label="Test set",marker='^',s=60)
axes[0].legend()
axes[0].set_title("Unscaled")


axes[1].scatter(X_train_scaled[:,0],X_train_scaled[:,1],c=y_train,label="Scaled Training set",s=60)
axes[1].scatter(X_test_scaled[:,0],X_test_scaled[:,1],c=y_test,label="Scaled Test set",marker='^',s=60)
axes[1].legend()
axes[1].set_title("Scaled")
```

```python
svm = SVC(C=100)
svm.fit(X_train, y_train)
print("Unscaled train score:{}".format(svm.score(X_train,y_train)))
print("Unscaled test score:{}".format(svm.score(X_test_scaled,y_test)))

svm.fit(X_train_scaled,y_train)
print("Scaled train score:{}".format(svm.score(X_train_scaled,y_train)))
print("Scaled test score:{}".format(svm.score(X_test_scaled,y_test)))
## Much better test score after scaling!!!!
```
