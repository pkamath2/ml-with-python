```python
import sklearn
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import mglearn
```

```python
cancer = sklearn.datasets.load_breast_cancer()
cancer_data_frame = pd.DataFrame(cancer.data, columns=cancer["feature_names"])
print(cancer)
print(cancer_data_frame[:100])
print(cancer.keys())
```

```python
#matrix = pd.scatter_matrix(cancer_data_frame[:100],c=cancer.target[:100], figsize=(30,30),marker='o',hist_kwds={'bins':20},s=60,alpha=.8,cmap=mglearn.cm3)
```

```python
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(cancer['data'],cancer['target'],stratify=cancer['target'],random_state=0)
```

```python
training_accuracy = []
test_accuracy = []
neighbors = range(1,11)
```

```python
for neighbor in neighbors:
    classifier = sklearn.neighbors.KNeighborsClassifier(n_neighbors=neighbor)
    classifier.fit(X_train, y_train)
    training_accuracy.append(classifier.score(X_train,y_train))
    test_accuracy.append(classifier.score(X_test,y_test))

print(neighbors)
print(training_accuracy)
print(test_accuracy)        
```

```python
plt.plot(neighbors,training_accuracy,label="training accuracy")
plt.plot(neighbors,test_accuracy,label="testing accuracy")
plt.ylabel("Accuracy")
plt.xlabel("neighbors")
plt.legend()
```
