```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import scipy.sparse as sparse
import mglearn
```

```python
sparse_matrix = sparse.csr_matrix(np.eye(4))
print(sparse_matrix)
```

```python
iris_dataset = sklearn.datasets.load_iris()
print(iris_dataset)
```

```python
print("type of data:\n {}".format(type(iris_dataset['data'])))
```

```python
print("shape of data:\n {}".format(iris_dataset['data'].shape))
```

```python
print("First 5 rows:\n{}".format(iris_dataset['data'][:5]))
```

```python
X_train,X_test,y_train,y_test = train_test_split(iris_dataset['data'],iris_dataset['target'],random_state=0)
print("Data training:\n{}".format(X_train))
print("Data test:\n{}".format(X_test))
print("Label training:\n{}".format(y_train))
print("Label test:\n{}".format(y_test))
```

```python
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset['feature_names'])
```

```python
grr = pd.scatter_matrix(iris_dataframe,c=y_train, figsize=(15,15),marker='o',hist_kwds={'bins':20},s=60,alpha=.8,cmap=mglearn.cm3)
```

```python
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
```

```python
print(knn)
```

```python
X_new = np.array([[5,2.9,1,0.2]])
prediction = knn.predict(X_new)
```

```python
print("Prediction:\n{}".format(prediction))
```

```python
y_pred=knn.predict(X_test)
print("Prediction:\n{}".format(y_pred))
```

```python
score = knn.score(X_test,y_test)
print("Score:\n{}".format(score))
```
