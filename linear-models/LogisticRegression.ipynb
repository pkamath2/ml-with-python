```python
import sklearn
%matplotlib inline
import matplotlib.pyplot as plt
import mglearn
import pandas as pd
```

```python
X,y=mglearn.datasets.make_forge()
X_train,X_test,y_train,y_test=sklearn.model_selection.train_test_split(X,y)
print(X_train)
```

```python
df = pd.DataFrame(X_train,columns=[["feature0","feature1"]])
```

```python
scatter_matrix = pd.scatter_matrix(df,c=y_train,figsize=(15,15),marker='o',hist_kwds={'bins':20},s=60,alpha=.8,cmap=mglearn.cm3)
```

```python
classifier = sklearn.linear_model.LogisticRegression()
classifier.fit(X_train,y_train)
print("Training score:\n{}".format(classifier.score(X_train,y_train)))
print("Testing score:\n{}".format(classifier.score(X_test,y_test)))
```

```python
classifier = sklearn.linear_model.LogisticRegression(C=.001)
classifier.fit(X_train,y_train)
print("Training score:\n{}".format(classifier.score(X_train,y_train)))
print("Testing score:\n{}".format(classifier.score(X_test,y_test)))
```
