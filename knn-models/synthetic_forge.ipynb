```python
import mglearn
import sklearn
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
X, y = mglearn.datasets.make_forge()
print("Classes:\n{}".format(y))
print("Dataset:\n{}".format(X))
```

```python
X_dataframe = pd.DataFrame(X, columns=['feature 1','feature 2'])
print("Dataframe:\n{}".format(X_dataframe))
```

```python
mt = pd.scatter_matrix(X_dataframe,c=y,marker='o',figsize=(10,10),s=60,alpha=.8,cmap=mglearn.cm3)
```

```python
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X,y,random_state=0)
```

```python
print("Training dataset:\n{}".format(X_train))
print("Test dataset:\n{}".format(X_test))
```

```python
fig, axes = plt.subplots(1,3,figsize=(10,3))
for n_neighbors, ax in zip([1,5,9],axes):
    classifier = sklearn.neighbors.KNeighborsClassifier(n_neighbors=n_neighbors).fit(X_train, y_train)
    mglearn.plots.plot_2d_separator(classifier, X_test, fill=True, eps=0.5, ax=ax, alpha=.4)
    mglearn.discrete_scatter(X[:,0],X[:,1],y,ax=ax)
axes[0].legend(loc=3)
    
```

```python
#1 Neighbors
classifier = sklearn.neighbors.KNeighborsClassifier(n_neighbors=1)
classifier.fit(X_train, y_train)
print("Run tests:\n{}".format(classifier.predict(X_test)))
score = classifier.score(X_test, y_test)
print("Score:\n{}".format(score))
```

```python
#3 Neighbors
classifier = sklearn.neighbors.KNeighborsClassifier(n_neighbors=3)
classifier.fit(X_train, y_train)
print("Run tests:\n{}".format(classifier.predict(X_test)))
score = classifier.score(X_test, y_test)
print("Score:\n{}".format(score))
```

```python
#9 Neighbors
classifier = sklearn.neighbors.KNeighborsClassifier(n_neighbors=9)
classifier.fit(X_train, y_train)
print("Run tests:\n{}".format(classifier.predict(X_test)))
score = classifier.score(X_test, y_test)
print("Score:\n{}".format(score))
```
