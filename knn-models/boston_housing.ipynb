```python
import sklearn
import mglearn
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
boston_housing = sklearn.datasets.load_boston()
print("Dataset Keys:\n{}".format(boston_housing.keys()))
print("Dataset shape:\n{}".format(boston_housing['data'].shape))
print(boston_housing['target']*10)
```

```python
#print("Dataset DESCR:\n{}".format(boston_housing['DESCR']))
```

```python
X,y = mglearn.datasets.load_extended_boston()
print(X.shape)
```

```python
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X,y*10)
# There is an error is I dont multiply target by 10. Seems like for Classifiers target should not be a float point. 
```

```python
testing_accuracy = []
training_accuracy = []
neighbors = range(1,11)
for neighbor in neighbors:
    classifier = sklearn.neighbors.KNeighborsClassifier(n_neighbors=neighbor)
    classifier.fit(X_train, y_train)
    training_accuracy.append(classifier.score(X_train, y_train))
    testing_accuracy.append(classifier.score(X_test, y_test))
```

```python
plt.plot(neighbors,testing_accuracy, label="testing accuracy")
plt.plot(neighbors,training_accuracy, label="training accuracy")
plt.legend()
#The accuracy is weird!!
```

```python
#Now lets use KNeighborsRegression
testing_accuracy = []
training_accuracy = []
neighbors = range(1,11)
for neighbor in neighbors:
    classifier = sklearn.neighbors.KNeighborsRegressor(n_neighbors=neighbor)
    classifier.fit(X_train, y_train)
    training_accuracy.append(classifier.score(X_train, y_train))
    testing_accuracy.append(classifier.score(X_test, y_test))
```

```python
plt.plot(neighbors,testing_accuracy, label="testing accuracy")
plt.plot(neighbors,training_accuracy, label="training accuracy")
plt.legend()
#This is better than Classifier... WHY-OH-WHY??? 
```
